```python

# import module
import pandas as pd
  
# Read excel file
# and store into a DataFrame

df38 = pd.read_excel('Mobi_System_Data_2021-01.xlsx')
df39 = pd.read_excel('Mobi_System_Data_2021-02.xlsx')
df40 = pd.read_excel('Mobi_System_Data_2021-03.xlsx')
df41 = pd.read_excel('Mobi_System_Data_2021-04.xlsx')

  
# concat both DataFrame into a single DataFrame
df = pd.concat([df38, df39, df40, df41])
  
# Export Dataframe into Excel file
df.to_excel('Mobi_System_Data_2021.xlsx', index=False)
```
